**Étape 1 : Collecte des données**

```python
# ÉTAPE 1 : COLLECTE DEPUIS UN PDF

import PyPDF2

def extraire_texte_pdf(chemin_pdf):
    """Extrait le texte d'un fichier PDF"""
    texte = ""

    with open(chemin_pdf, 'rb') as fichier:
        lecteur = PyPDF2.PdfReader(fichier)

        for page in lecteur.pages:
            texte += page.extract_text()

    return texte

# Utilisation
chemin = "/content/ai-revolution.pdf"  # Remplacez par le chemin de votre PDF
texte_brut = extraire_texte_pdf(chemin)
print("\n" + "=" * 60)
print("📊 RÉSUMÉ DE LA COLLECTE DEPUIS PDF")
print("=" * 60)
print(f"Texte extrait : {len(texte_brut)} caractères")
print(texte_brut[:100])  # Aperçu
```

### **Étape 2 : Prétraitement (Nettoyage)**

- **Tokenisation** : Découper le texte en mots/phrases
- **Tokenisation** : Découper le texte en mots/phrases
    
- **Normalisation** : Mettre en minuscules, supprimer les accents, etc.
    
- **Suppression des stopwords** : Enlever les mots vides ("le", "la", "et", "de"...)
    
- **Stemming/Lemmatisation** : Ramener les mots à leur racine ("mangé" → "manger")

```python
# ============================================================================
# ÉTAPE 2 : PRÉTRAITEMENT (SANS SPACY - Solution manuelle)
# ============================================================================

import re
from nltk.corpus import stopwords
import nltk

# Télécharger les stopwords arabes nltk
nltk.download('stopwords')

# Liste manuelle des stopwords arabes (communs)
stopwords_arabes = set([
    'في', 'من', 'على', 'إلى', 'عن', 'مع', 'بين', 'بعد', 'قبل', 'تحت',
    'فوق', 'خلال', 'حول', 'حتى', 'عند', 'مثل', 'كان', 'هو', 'هي', 'هم',
    'أن', 'إن', 'أنه', 'أنها', 'هذا', 'هذه', 'ذلك', 'تلك', 'هؤلاء',
    'و', 'ف', 'ثم', 'أو', 'لكن', 'لذلك', 'لأن', 'لأنه', 'لأنها'
])

# ----------------------------------------------------------------------------
# 1. TOKENISATION (manuelle)
# ----------------------------------------------------------------------------
print("\n" + "=" * 60)
print("1. TOKENISATION")
print("=" * 60)

def tokeniser_manuel(texte):
    # Séparer par espaces et ponctuation
    tokens = re.findall(r'[\w\u0600-\u06FF]+', texte)
    return tokens

tokens = tokeniser_manuel(texte_brut)
print(f"Nombre de tokens : {len(tokens)}")
print(f"Aperçu : {tokens[:20]}")

# ----------------------------------------------------------------------------
# 2. NORMALISATION
# ----------------------------------------------------------------------------
print("\n" + "=" * 60)
print("2. NORMALISATION")
print("=" * 60)

def normaliser_arabe(texte):
    texte = texte.lower()
    # Supprimer diacritiques arabes (optionnel)
    texte = re.sub(r'[\u064B-\u0652]', '', texte)  # Enlever harakat
    # Garder lettres arabes et espaces
    texte = re.sub(r'[^\w\s\u0600-\u06FF]', '', texte)
    # Supprimer chiffres
    texte = re.sub(r'\d+', '', texte)
    return texte

texte_normalise = normaliser_arabe(texte_brut)
print(f"Texte normalisé (aperçu) : {texte_normalise[:200]}")

# ----------------------------------------------------------------------------
# 3. SUPPRESSION DES STOPWORDS
# ----------------------------------------------------------------------------
print("\n" + "=" * 60)
print("3. SUPPRESSION DES STOPWORDS")
print("=" * 60)

def supprimer_stopwords_manuel(tokens):
    return [token for token in tokens if token not in stopwords_arabes]

tokens_filtres = supprimer_stopwords_manuel(tokens[:100])
print(f"Tokens avant : {len(tokens[:100])}")
print(f"Tokens après : {len(tokens_filtres)}")
print(f"Aperçu : {tokens_filtres[:15]}")

# ----------------------------------------------------------------------------
# 4. STEMMING SIMPLE (sans lemmetisation complexe)
# ----------------------------------------------------------------------------
print("\n" + "=" * 60)
print("4. STEMMING SIMPLE")
print("=" * 60)

def stemming_simple(tokens):
    """Supprime les préfixes et suffixes communs en arabe"""
    stems = []
    for token in tokens:
        # Supprimer le 'ال' au début
        if token.startswith('ال'):
            token = token[2:]
        # Supprimer 'ة' à la fin
        if token.endswith('ة'):
            token = token[:-1]
        stems.append(token)
    return stems

stems = stemming_simple(tokens_filtres[:15])
print(f"Stems obtenus : {stems}")
print("\n✅ ÉTAPE 2 TERMINÉE (Version sans SpaCy)")
```

 **Étape 3 : Analyse syntaxique**

- **POS Tagging** : Identifier la nature des mots (nom, verbe, adjectif...)
    
- **Dependency Parsing** : Comprendre les relations entre mots (sujet, objet...)

```python
# ============================================================================
# ÉTAPE 3 : ANALYSE SYNTAXIQUE (VERSION SIMPLIFIÉE & ROBUSTE)
# ============================================================================

print("\n" + "=" * 60)
print("ÉTAPE 3 : ANALYSE SYNTAXIQUE")
print("=" * 60)

# ----------------------------------------------------------------------------
# 1. NORMALISATION SIMPLE (GARDER UNE SEULE VERSION)
# ----------------------------------------------------------------------------

def normaliser(token):
    token = token.strip()

    mapping = {
        'ى': 'ي',
        'أ': 'ا',
        'إ': 'ا',
        'آ': 'ا',
        'ة': 'ه',
        'ؤ': 'و',
        'ئ': 'ي',
    }

    for k, v in mapping.items():
        token = token.replace(k, v)

    return token


# ----------------------------------------------------------------------------
# 2. DICTIONNAIRES SIMPLIFIÉS
# ----------------------------------------------------------------------------

PREPOSITIONS = {'في', 'من', 'على', 'إلى', 'عن', 'مع', 'بعد', 'قبل', 'حتى'}
CONJ = {'و', 'ف', 'ثم', 'أو', 'لكن', 'لأن'}
PRON = {'هو', 'هي', 'هم', 'أنا', 'نحن', 'أنت'}
VERB_HINTS = {'كان', 'يكون', 'يعمل', 'يستخدم', 'أصبح', 'أدى'}


# ----------------------------------------------------------------------------
# 3. POS TAGGING SIMPLIFIÉ (RULE-BASED)
# ----------------------------------------------------------------------------

def pos_tag(token):
    t = normaliser(token)

    # Préposition
    if t in PREPOSITIONS:
        return "ADP"

    # Conjonction
    if t in CONJ:
        return "CONJ"

    # Pronom
    if t in PRON:
        return "PRON"

    # Verbe (heuristique)
    if t in VERB_HINTS or t.endswith('ت') or t.startswith('ي'):
        return "VERB"

    # Adjectif heuristique
    if t.endswith('ي') or t.endswith('ة'):
        return "ADJ"

    # Sinon nom
    return "NOUN"


# ----------------------------------------------------------------------------
# 4. ANALYSE DES TOKENS
# ----------------------------------------------------------------------------

def analyser_tokens(tokens, limit=30):
    results = []

    stats = {
        "VERB": 0,
        "NOUN": 0,
        "ADJ": 0,
        "ADP": 0,
        "CONJ": 0,
        "PRON": 0
    }

    print("\nAnalyse:")
    print("-" * 60)

    for token in tokens[:limit]:
        norm = normaliser(token)
        pos = pos_tag(token)

        stats[pos] += 1
        results.append((token, norm, pos))

        emoji = {
            "VERB": "🔵",
            "NOUN": "🟢",
            "ADJ": "🟡",
            "ADP": "⚪",
            "CONJ": "🟠",
            "PRON": "🟣"
        }.get(pos, "⚫")

        print(f"{emoji} {token:15} → {norm:15} → {pos}")

    return stats


# ----------------------------------------------------------------------------
# 5. UTILISATION AVEC TON PIPELINE EXISTANT
# ----------------------------------------------------------------------------

# ⚠️ utiliser les tokens de ton ÉTAPE 2
stats = analyser_tokens(tokens, 30)


# ----------------------------------------------------------------------------
# 6. STATISTIQUES
# ----------------------------------------------------------------------------

print("\n" + "=" * 60)
print("📊 STATISTIQUES")
print("=" * 60)

total = sum(stats.values())

for k, v in sorted(stats.items(), key=lambda x: x[1], reverse=True):
    print(f"{k:5} : {v} ({(v/total)*100:.1f}%)")


print("\n✅ ÉTAPE 3 TERMINÉE")
```

###  **Étape 4 : Analyse sémantique**

- **NER** (Named Entity Recognition) : Identifier les entités nommées
    
    - Personnes ("Steve Jobs")
        
    - Lieux ("Californie")
        
    - Organisations ("Apple")
        
    - Dates, montants, etc.

```python
# ============================================================================
# ÉTAPE 4 : ANALYSE SÉMANTIQUE - NER (Reconnaissance d'Entités Nommées)
# ============================================================================

print("\n" + "=" * 60)
print("ÉTAPE 4 : ANALYSE SÉMANTIQUE (NER)")
print("=" * 60)

# ----------------------------------------------------------------------------
# 1. DICTIONNAIRES D'ENTITÉS POUR L'ARABE
# ----------------------------------------------------------------------------

# Personnes célèbres / noms propres
PERSONNES = set([
    'احمد', 'محمد', 'علي', 'حسن', 'حسين', 'فاطمة', 'مريم', 'سارة',
    'يوسف', 'ابراهيم', 'موسى', 'عيسى', 'خديجة', 'عائشة', 'بلال',
    'عمر', 'عثمان', 'طلحة', 'الزبير', 'سعد', 'زيد', 'عمرو'
])

# Lieux (pays, villes, régions)
LIEUX = set([
    'المغرب', 'الدار البيضاء', 'الرباط', 'فاس', 'مراكش', 'طنجة', 'أكادير',
    'مصر', 'القاهرة', 'السعودية', 'الرياض', 'جدة', 'مكة', 'المدينة',
    'الجزائر', 'تونس', 'ليبيا', 'السودان', 'العراق', 'سوريا', 'لبنان',
    'فلسطين', 'الأردن', 'الكويت', 'قطر', 'البحرين', 'عمان', 'اليمن',
    'فرنسا', 'باريس', 'أمريكا', 'واشنطن', 'لندن', 'برلين', 'الصين', 'طوكيو'
])

# Organisations
ORGANISATIONS = set([
    'جامعة', 'الجامعة', 'مدرسة', 'شركة', 'مستشفى', 'وزارة', 'بلدية',
    'الأمم المتحدة', 'اليونسكو', 'الناتو', 'الاتحاد الأوروبي', 'منظمة الصحة العالمية',
    'فاس', 'الرباط', 'الدار البيضاء'
])

# Dates (mots indicateurs)
DATES_INDICATEURS = set([
    'اليوم', 'غدا', 'أمس', 'الأسبوع', 'الشهر', 'السنة', 'القرن',
    'يناير', 'فبراير', 'مارس', 'أبريل', 'ماي', 'يونيو', 'يوليوز',
    'غشت', 'شتنبر', 'أكتوبر', 'نونبر', 'دجنبر', 'الاثنين', 'الثلاثاء'
])

# Technologies / IA (domaine spécifique)
TECHNOLOGIES = set([
    'الذكاء الاصطناعي', 'تعلم الآلة', 'تعلم عميق', 'بيانات ضخمة', 'خوارزميات',
    'روبوت', 'شبكات عصبية', 'معالجة اللغة', 'رؤية حاسوبية', 'إنترنت الأشياء'
])

# ----------------------------------------------------------------------------
# 2. FONCTION DE NER BASÉE SUR DES RÈGLES
# ----------------------------------------------------------------------------

def detecter_entites(tokens):
    """
    Détecte les entités nommées dans les tokens
    Retourne une liste de (entité, type)
    """
    entites = []

    i = 0
    while i < len(tokens):
        token = tokens[i]
        token_normalise = normaliser(token) if 'normaliser' in dir() else token

        # Vérifier les entités multi-mots (jusqu'à 3 mots)
        if i < len(tokens) - 1:
            bigramme = token + " " + tokens[i + 1]
            trigramme = token + " " + tokens[i + 1] + " " + tokens[i + 2] if i < len(tokens) - 2 else ""

            # Technologie (peut être plusieurs mots)
            if trigramme in TECHNOLOGIES:
                entites.append((trigramme, "TECH"))
                i += 3
                continue
            elif bigramme in TECHNOLOGIES:
                entites.append((bigramme, "TECH"))
                i += 2
                continue
            elif token in TECHNOLOGIES:
                entites.append((token, "TECH"))
                i += 1
                continue

        # Vérifier chaque catégorie
        if token in PERSONNES:
            entites.append((token, "PERSON"))
        elif token in LIEUX:
            entites.append((token, "LOC"))
        elif token in ORGANISATIONS:
            entites.append((token, "ORG"))
        elif token in DATES_INDICATEURS:
            entites.append((token, "DATE"))
        elif token in TECHNOLOGIES:
            entites.append((token, "TECH"))

        # Heuristiques pour noms propres (commence par lettre majuscule visuelle)
        elif token.startswith('ا') and len(token) > 3 and token not in stopwords_arabes:
            # Peut être un nom propre
            if token not in LIEUX and token not in PERSONNES:
                entites.append((token, "MISC"))  # Autre entité
        else:
            # Pas une entité
            pass

        i += 1

    return entites

# ----------------------------------------------------------------------------
# 3. NER AVEC CONTEXTE (VERSION AMÉLIORÉE)
# ----------------------------------------------------------------------------

class NER_arabe:
    """Classe pour la reconnaissance d'entités nommées en arabe"""

    def __init__(self):
        self.personnes = PERSONNES
        self.lieux = LIEUX
        self.organisations = ORGANISATIONS
        self.dates = DATES_INDICATEURS
        self.technologies = TECHNOLOGIES

    def analyser(self, tokens):
        """Analyse les tokens et retourne les entités avec leur contexte"""
        entites = []

        for i, token in enumerate(tokens):
            token_norm = normaliser(token) if 'normaliser' in dir() else token
            context = self._get_context(tokens, i)

            # Détection par type
            entite_type = self._detecter_type(token_norm)

            if entite_type:
                entites.append({
                    'entite': token,
                    'type': entite_type,
                    'position': i,
                    'contexte': context
                })

        return entites

    def _get_context(self, tokens, position, window=2):
        """Retourne le contexte autour de l'entité"""
        debut = max(0, position - window)
        fin = min(len(tokens), position + window + 1)
        return " ".join(tokens[debut:fin])

    def _detecter_type(self, token):
        """Détecte le type d'entité"""
        if token in self.personnes:
            return "PERSONNE"
        elif token in self.lieux:
            return "LIEU"
        elif token in self.organisations:
            return "ORGANISATION"
        elif token in self.dates:
            return "DATE"
        elif token in self.technologies:
            return "TECHNOLOGIE"

        # Heuristiques
        if token.endswith('ة') and len(token) > 4:
            return "POSSIBLE_NOM"

        return None

# ----------------------------------------------------------------------------
# 4. VISUALISATION DES ENTITÉS DÉTECTÉES
# ----------------------------------------------------------------------------

def visualiser_entites(texte, entites):
    """Affiche le texte avec les entités colorées visuellement"""
    print("\n📝 Texte avec entités détectées:")
    print("-" * 60)

    # Couleurs pour l'affichage (symboles)
    couleurs = {
        'PERSON': '👤',
        'PERSONNE': '👤',
        'LOC': '📍',
        'LIEU': '📍',
        'ORG': '🏢',
        'ORGANISATION': '🏢',
        'DATE': '📅',
        'TECH': '💻',
        'TECHNOLOGIE': '💻',
        'MISC': '🏷️'
    }

    for entite, type_entite in entites:
        emoji = couleurs.get(type_entite, '🔹')
        print(f"{emoji} {type_entite:12} : {entite}")

# ----------------------------------------------------------------------------
# 5. ANALYSE DU TEXTE
# ----------------------------------------------------------------------------

print("\n" + "=" * 60)
print("🔍 DÉTECTION DES ENTITÉS NOMMÉES")
print("=" * 60)

# Utiliser les tokens de l'étape 2
print(f"\nAnalyse de {min(len(tokens), 50)} tokens...")

# Détecter les entités
ner = NER_arabe()
entites_trouvees = ner.analyser(tokens[:100])

if entites_trouvees:
    print(f"\n✅ {len(entites_trouvees)} entités trouvées:")
    print("-" * 60)

    for entite in entites_trouvees:
        print(f"  {entite['type']:12} : {entite['entite']}")
        print(f"    → Contexte: \"{entite['contexte']}\"")
        print()
else:
    print("\n⚠️ Aucune entité nommée détectée dans les 100 premiers tokens")
    print("   (C'est normal si le texte contient peu de noms propres)")

# ----------------------------------------------------------------------------
# 6. STATISTIQUES DES ENTITÉS
# ----------------------------------------------------------------------------
print("\n" + "=" * 60)
print("📊 STATISTIQUES DES ENTITÉS")
print("=" * 60)

# Analyser tout le texte (limité pour performance)
entites_total = ner.analyser(tokens[:200])

stats_entites = {}
for entite in entites_total:
    type_ent = entite['type']
    stats_entites[type_ent] = stats_entites.get(type_ent, 0) + 1

if stats_entites:
    print("\nDistribution des types d'entités:")
    for type_ent, count in sorted(stats_entites.items(), key=lambda x: x[1], reverse=True):
        barre = "█" * min(count, 20)
        print(f"  {type_ent:12} : {count:3} {barre}")
else:
    print("\nAucune entité trouvée dans cette portion de texte.")
```

 **Étape 5 : Vectorisation (Représentation numérique)**

- Convertir les mots en nombres
    
- **TF-IDF** : Mesurer l'importance des mots
    
- **Word Embeddings** : Vecteurs contextuels (Word2Vec, BERT...)

```python
# ============================================================================
# ÉTAPE 5 : VECTORISATION (REPRÉSENTATION NUMÉRIQUE)
# ============================================================================

print("\n" + "=" * 60)
print("ÉTAPE 5 : VECTORISATION (TF-IDF)")
print("=" * 60)

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np

# ----------------------------------------------------------------------------
# 1. PRÉPARATION DU CORPUS (ensemble de documents)
# ----------------------------------------------------------------------------

print("\n📝 1. Préparation du corpus...")

# Utiliser les tokens nettoyés (sans stopwords) pour construire des documents
# On va créer plusieurs "documents" à partir des chunks du texte

def creer_documents(tokens, taille_doc=100):
    """Crée des documents en regroupant les tokens par paquets"""
    docs = []
    for i in range(0, len(tokens), taille_doc):
        chunk = tokens[i:i+taille_doc]
        doc = " ".join(chunk)
        docs.append(doc)
    return docs

# Créer des documents à partir des tokens
documents = creer_documents(tokens, 100)
print(f"✅ {len(documents)} documents créés (100 tokens par document)")

# Afficher un aperçu du premier document
if documents:
    print(f"\nAperçu du document 1:\n{documents[0][:200]}...")

# ----------------------------------------------------------------------------
# 2. VECTORISATION TF-IDF
# ----------------------------------------------------------------------------

print("\n" + "=" * 60)
print("🔢 2. Calcul TF-IDF")
print("=" * 60)

def vectoriser_tfidf(documents, max_features=100):
    """
    Convertit les documents en vecteurs TF-IDF

    TF = Term Frequency (fréquence du mot dans le document)
    IDF = Inverse Document Frequency (rareté du mot dans tous les documents)
    """

    # Créer le vectoriseur TF-IDF
    vectorizer = TfidfVectorizer(
        max_features=max_features,  # Nombre max de mots à considérer
        token_pattern=r'[\w\u0600-\u06FF]+',  # Reconnaître les caractères arabes
        ngram_range=(1, 2)  # Unigrammes et bigrammes
    )

    # Transformer les documents en matrice
    tfidf_matrix = vectorizer.fit_transform(documents)

    print(f"✅ Matrice TF-IDF créée: {tfidf_matrix.shape}")
    print(f"   - {tfidf_matrix.shape[0]} documents")
    print(f"   - {tfidf_matrix.shape[1]} features (mots)")

    return tfidf_matrix, vectorizer

# Appliquer TF-IDF
tfidf_matrix, vectorizer = vectoriser_tfidf(documents)

# ----------------------------------------------------------------------------
# 3. AFFICHAGE DES RÉSULTATS TF-IDF
# ----------------------------------------------------------------------------

print("\n" + "=" * 60)
print("📊 3. Visualisation des vecteurs TF-IDF")
print("=" * 60)

# Récupérer les noms des features (mots)
feature_names = vectorizer.get_feature_names_out()

# Créer un DataFrame pour visualiser
df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=feature_names,
    index=[f"Doc_{i+1}" for i in range(len(documents))]
)

print("\n📋 Premiers documents (5 premiers mots):")
print(df_tfidf.iloc[:2, :5])

# ----------------------------------------------------------------------------
# 4. MOTS LES PLUS IMPORTANTS PAR DOCUMENT
# ----------------------------------------------------------------------------

print("\n" + "=" * 60)
print("⭐ 4. Mots les plus importants par document")
print("=" * 60)

def mots_importants_par_doc(tfidf_matrix, feature_names, doc_index, top_k=10):
    """Affiche les top-k mots les plus importants pour un document"""

    # Récupérer les scores du document
    doc_scores = tfidf_matrix[doc_index].toarray()[0]

    # Créer une liste de (mot, score)
    mots_scores = list(zip(feature_names, doc_scores))

    # Trier par score décroissant
    mots_scores_tries = sorted(mots_scores, key=lambda x: x[1], reverse=True)

    # Afficher les top-k
    print(f"\n📄 Document {doc_index + 1}:")
    print("-" * 40)
    for i, (mot, score) in enumerate(mots_scores_tries[:top_k], 1):
        if score > 0:
            print(f"  {i:2}. {mot:20} : {score:.4f}")

    return mots_scores_tries[:top_k]

# Afficher les mots importants pour les 3 premiers documents
for i in range(min(3, len(documents))):
    mots_importants_par_doc(tfidf_matrix, feature_names, i, 8)

# ----------------------------------------------------------------------------
# 5. SIMILARITÉ ENTRE DOCUMENTS
# ----------------------------------------------------------------------------

print("\n" + "=" * 60)
print("🔗 5. Similarité entre documents (cosinus)")
print("=" * 60)

from sklearn.metrics.pairwise import cosine_similarity

def calculer_similarite(tfidf_matrix):
    """Calcule la similarité cosinus entre tous les documents"""

    # Calculer la matrice de similarité
    similarity_matrix = cosine_similarity(tfidf_matrix)

    # Créer un DataFrame pour visualisation
    df_similarity = pd.DataFrame(
        similarity_matrix,
        index=[f"Doc_{i+1}" for i in range(len(documents))],
        columns=[f"Doc_{i+1}" for i in range(len(documents))]
    )

    return df_similarity

# Calculer la similarité
df_similarity = calculer_similarite(tfidf_matrix)

print("\n📊 Matrice de similarité (5 premiers documents):")
print(df_similarity.iloc[:5, :5].round(3))

# Afficher les documents les plus similaires au premier document
if len(documents) > 1:
    print(f"\n🔍 Documents les plus similaires au Document 1:")
    similarities = df_similarity.iloc[0].sort_values(ascending=False)
    for doc_name, score in similarities.iloc[1:4].items():
        print(f"   {doc_name}: {score:.3f}")

# ----------------------------------------------------------------------------
# 6. VECTEUR POUR UNE PHRASE SPÉCIFIQUE
# ----------------------------------------------------------------------------

print("\n" + "=" * 60)
print("🎯 6. Vectorisation d'une nouvelle phrase")
print("=" * 60)

def vectoriser_phrase(phrase, vectorizer):
    """Vectorise une nouvelle phrase en utilisant le vectorizer existant"""

    # Nettoyer la phrase
    phrase_clean = normaliser_arabe(phrase) if 'normaliser_arabe' in dir() else phrase.lower()

    # Vectoriser
    vecteur = vectorizer.transform([phrase_clean])

    return vecteur

# Tester avec une phrase exemple
phrase_test = "الذكاء الاصطناعي سيقوم بتغيير العالم في المستقبل"
print(f"Phrase test: {phrase_test}")

vecteur_phrase = vectoriser_phrase(phrase_test, vectorizer)
print(f"\n✅ Vecteur créé: {vecteur_phrase.shape}")
print(f"   Nombre de features non nulles: {vecteur_phrase.nnz}")

# ----------------------------------------------------------------------------
# 7. STATISTIQUES DES VECTEURS
# ----------------------------------------------------------------------------

print("\n" + "=" * 60)
print("📈 7. Statistiques des vecteurs")
print("=" * 60)

def stats_vectorisation(tfidf_matrix, feature_names):
    """Affiche des statistiques sur la vectorisation"""

    # Statistiques globales
    non_zero_ratio = (tfidf_matrix > 0).sum() / (tfidf_matrix.shape[0] * tfidf_matrix.shape[1])

    # Moyenne des scores par document
    moyennes_par_doc = tfidf_matrix.mean(axis=1).A1

    # Scores maximaux par document
    max_scores = tfidf_matrix.max(axis=1).toarray().flatten()

    print(f"""
    📊 Statistiques de vectorisation:

    Nombre de documents        : {tfidf_matrix.shape[0]}
    Nombre de features (mots)  : {tfidf_matrix.shape[1]}
    Taux de remplissage        : {non_zero_ratio*100:.2f}%
    Score TF-IDF moyen         : {tfidf_matrix.mean():.4f}
    Score TF-IDF max global    : {tfidf_matrix.max():.4f}
    """)

    # Mot le plus important globalement
    scores_par_mot = tfidf_matrix.sum(axis=0).A1
    mot_plus_important = feature_names[scores_par_mot.argmax()]
    score_max = scores_par_mot.max()

    print(f"🏆 Mot le plus important globalement: '{mot_plus_important}' (score total: {score_max:.2f})")

    return moyennes_par_doc

moyennes = stats_vectorisation(tfidf_matrix, feature_names)

# ----------------------------------------------------------------------------
# 8. EXPORT DES VECTEURS (optionnel)
# ----------------------------------------------------------------------------

print("\n" + "=" * 60)
print("💾 8. Sauvegarde des résultats (optionnel)")
print("=" * 60)

def sauvegarder_resultats():
    """Sauvegarde les matrices vectorisées pour utilisation future"""

    # Sauvegarder la matrice TF-IDF
    import pickle

    try:
        # Sauvegarder le vectorizer
        with open('tfidf_vectorizer.pkl', 'wb') as f:
            pickle.dump(vectorizer, f)

        # Sauvegarder la documentation
        print("✅ Vectorizer sauvegardé dans 'tfidf_vectorizer.pkl'")
        print("   Utilisation future: vectorizer = pickle.load(open('tfidf_vectorizer.pkl', 'rb'))")
    except Exception as e:
        print(f"⚠️ Sauvegarde non disponible: {e}")

# Optionnel: décommenter pour sauvegarder
# sauvegarder_resultats()

# ----------------------------------------------------------------------------
# 9. RÉSUMÉ FINAL
# ----------------------------------------------------------------------------

print("\n" + "=" * 60)
print("🎯 RÉSUMÉ DE L'ÉTAPE 5")
print("=" * 60)

print(f"""
✅ Vectorisation terminée avec succès!

Résultats obtenus:
- {tfidf_matrix.shape[0]} documents vectorisés
- {tfidf_matrix.shape[1]} mots uniques dans le vocabulaire
- Matrice TF-IDF de taille {tfidf_matrix.shape}

Exemple de vecteur (Document 1):
""")

# Afficher un exemple de vecteur
if len(documents) > 0:
    doc1_vector = tfidf_matrix[0].toarray()[0]
    non_zero_indices = doc1_vector.nonzero()[0]
    print(f"  - {len(non_zero_indices)} mots importants sur {len(feature_names)}")
    print(f"  - Exemple: {feature_names[non_zero_indices[0]]} = {doc1_vector[non_zero_indices[0]]:.3f}")

print("\n" + "=" * 60)
print("🎉 PIPELINE NLP COMPLET - 100% TERMINÉ! 🎉")
print("=" * 60)

print("""
Étapes complétées:
✅ Étape 1: Collecte depuis PDF
✅ Étape 2: Prétraitement (Tokenisation, Stopwords, Stemming)
✅ Étape 3: Analyse syntaxique (POS Tagging)
✅ Étape 4: Analyse sémantique (NER)
✅ Étape 5: Vectorisation (TF-IDF)
""")
```
